## US Sub National Resilience - Data Cleaning Pipeline


```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
# Creating a dictionary to merge the csv's
my_dict = {
    'df_ep_com.csv': 'electricity_prices_commercial',
    'df_ep_ind.csv': 'electricity_prices_industrial',
    'df_ep_res.csv': 'electricity_prices_residential',
    'df_ep_tra.csvgit': 'electricity_prices_transport',
    'df_ng_1.csv': 'natural_gas_prices_2023_2026',
    'df_ng_2.csv': 'natural_gas_prices_2022_2023',
    'df_ng_demand_1.csv' : 'natural_gas_demand_2023_2026',
    'df_ng_demand_2.csv' : 'natural_gas_demand_2022_2023',
    'Quartely_GDP1__ALL_AREAS_2005_2025.csv': "quart_gdp_state",
    'Fossil_Fuel_Industry_GDP_State_Wise.csv': "fossil_fuel_gdp_state",
    'Size_Production_Fossil_Fuels_State_Wise.xlsx':'fossil_fuel_prod_state',
    'Pce_state_wise_dollar'
}

with pd.ExcelWriter('cleaned_data.xlsx') as writer:
    for file_name, sheet_name in my_dict.items():
        # 1. Read the data correctly depending on file format
        if file_name.endswith('.csv'):  # Pro-tip: adding the dot protects against name bugs
            df = pd.read_csv(file_name)
        else:
            df = pd.read_excel(file_name)
            
        # 2. Write EVERY parsed dataframe to a sheet (moved outside the else block)
        df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"Saved {file_name} to sheet {sheet_name}")

print("All files have been processed and saved to cleaned_data.xlsx")
```
